```python
import pandas as pd

def reemplazar_abstain(df: pd.DataFrame) -> pd.DataFrame:
    p_cols = ['p_1', 'p_2', 'p_3', 'p_4', 'p_5']

    # ndice 1..5 del p_i mximo (vectorizado)
    i_max = df[p_cols].values.argmax(axis=1) + 1

    # mask de filas a reemplazar (case-insensitive, por si acaso)
    mask = df['pred'].astype(str).str.lower().eq('abstain')

    # asigna usando el mismo mask en ambos lados (evita NaN por desalineacin)
    df.loc[mask, 'pred'] = i_max[mask]

    # opcional: si quieres que 'pred' sea entero luego del reemplazo:
    # si hay mezcla (strings y enteros), esto lo pasa a tipo 'Int64' permisivo
    try:
        df['pred'] = pd.to_numeric(df['pred'], errors='raise').astype('Int64')
    except Exception:
        # si prefieres mantener cadenas, comenta el bloque try/except
        pass

    return df


open_smile_full = reemplazar_abstain (open_smile)

open_smile_full
```
